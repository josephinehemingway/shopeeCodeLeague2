```python
import string

#Download stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def clean_data(data):
    #remove punctuations
    nopunc= [char for char in data if char not in string.punctuation]
    nopunc=''.join(nopunc)
    
    clean_words = [word for word in nopunc.split() if word.lower() not in stopwords.words('indonesian')]

    #return cleanted text
    return (clean_words)
```

```python
#lowercase 
POI_df= POI_table["raw_address"].str.lower()
POI_df= POI_table["raw_address"]
```

```python
POI_df
```

```python
#cleaned data
cleanPOI= POI_df.apply(clean_data)
```

# Preprocessing: Tokenisation of words

Tokenisation is the process to assign each word in the messages a vector so that it could be analysed. 

This is achieved by applying `CountVectorizer()` to count the number of occurrences of each word in each adress, resulting in a sparse array with the rows corresponding to index of address and columns corresponding to the different vocabulary.

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(min_df=0.0)
bow_transformer
```

```python
cleanPOI[299998]
```

```python
bow_matrix = bow_transformer.fit_transform(cleanPOI)
bow_matrix = bow_matrix.toarray()
```

```python
#ncounting number of different words in address
vocab= bow_transformer.get_feature_names()
```

```python
add_matrix=pd.DataFrame(bow_matrix,columns=vocab)
```
