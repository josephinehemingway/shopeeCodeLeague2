```python
import json
import pandas as pd
import numpy as np # linear algebra
import re
with open('street_start_dict.json', 'r') as fp:
    street_start_dict = json.load(fp)
with open('street_end_dict.json', 'r') as fp:
    street_end_dict = json.load(fp)
with open('poi_start_dict.json', 'r') as fp:
    poi_start_dict = json.load(fp)
with open('poi_end_dict.json', 'r') as fp:
    poi_end_dict = json.load(fp)
    
df = pd.read_csv('test.csv')

for index, row in df.iterrows():
    address=row['raw_address']
#   ASSUMPTION #1 EVERYTHING AFTER LAST OCCURANCE OF ',' IS USELESS
    comma_index = next((i for i, x in enumerate(address) if x == ','), None)
    address= address[:comma_index]
#   ASSUMPTION #2 EVERYTHING AFTER POSTAL CODE OCCURANCE IS USELESS
    indexOfPostalCode= re.search(r"\D(\d{5})\D", address)
    if indexOfPostalCode!=None:
        address = address[:indexOfPostalCode.start()]
#   Remove trailing whitespace and commas
    df.at[index,'raw_address'] = address.strip(' ,')

df
```

```python
df['guessed_street'] = ''
df['guessed_poi'] = ''
for index, row in df.iterrows():
    address= row['raw_address']
#     Street assignment
    address_words=address.split(' ')
    streetPreIndex=0
    streetPostIndex=len(address)-1
    for key in street_start_dict:
        try:
            street_index= address_words.index(key)
            streetPreIndex=street_index
            break
        except ValueError:
            pass 
    address_words=address_words[streetPreIndex:]
    for key in street_end_dict:
        try:
            street_index = address_words.index(key)
            streetPostIndex=street_index
            break
        except ValueError:
            pass 
    df.at[index,'guessed_street']=' '.join(address_words[streetPreIndex:streetPostIndex])

    #     poi assignment
    address_words=address.split(' ')
    poiPreIndex=0
    poiPostIndex=len(address)-1
    for key in poi_start_dict:
        try:
            poi_index= address_words.index(key)
            poiPreIndex=poi_index
            break
        except ValueError:
            pass 
    address_words=address_words[poiPreIndex:]
    for key in poi_end_dict:
        try:
            poi_index = address_words.index(key)
            poiPostIndex=poi_index
            break
        except ValueError:
            pass 
    df.at[index,'guessed_poi']=' '.join(address_words[poiPreIndex:poiPostIndex])
df
```

```python
import pickle
# unused =0 poi= 1 street =2

text_clf = pickle.load(open('NaiveBayes.pickle', 'rb'))

predictedStreetProba = text_clf.predict_proba(df.guessed_street)
df['predictedStreetProba']=[item[2] for item in predictedStreetProba.tolist()] 

predictedPoiProba = text_clf.predict_proba(df.guessed_poi)
df['predictedPoiProba']=[item[1] for item in predictedPoiProba.tolist()] 
df

for index, row in df.iterrows():
    if row['predictedStreetProba']<0.65:
        df.at[index,'guessed_street'] = ''
    if row['predictedPoiProba']<0.65:
        df.at[index,'guessed_poi'] = ''        

df['POI/street']= df["guessed_poi"]+"/"+df["guessed_street"]
df
```

```python
submission = df.drop(columns=['raw_address', 'guessed_street','guessed_poi','predictedStreetProba','predictedPoiProba'])
submission
submission.to_csv('submission.csv', index=False)
```
