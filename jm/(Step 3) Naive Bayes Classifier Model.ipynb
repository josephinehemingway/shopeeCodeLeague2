```python
# IMPORT AND INITIALIZE
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

df = pd.read_csv('train_test_set.csv')
df = df.replace(np.nan, '', regex=True)

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python


text_clf = Pipeline([('vect', CountVectorizer(ngram_range=(1, 2))),('tfidf', TfidfTransformer(use_idf=False)),('clf', MultinomialNB(alpha=0.01)),])


text_clf = text_clf.fit(train.text, train.category)
```

```python
predicted = text_clf.predict(test.text)
np.mean(predicted == test.category)
```

```python
predicted = text_clf.predict(df.text)
predicted_proba= text_clf.predict_proba(df.text)
np.mean(predicted == df.category)
df['predicted'] = predicted.tolist()
df['predicted_proba'] = predicted_proba.tolist()
for index, row in df.iterrows():
    df.at[index,'predicted_proba'] = round(row['predicted_proba'][row['predicted']],3)
print(df)

# Save classifier model
import pickle
f = open('NaiveBayes.pickle', 'wb')
pickle.dump(text_clf, f)
f.close()
```
